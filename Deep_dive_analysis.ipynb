```python
# TASK 3 - Deep Dive Analysis & Dashboard Preparation

# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Step 2: Load Dataset
df = pd.read_csv("cleaned_dataset.csv")

print("Dataset Preview")
print(df.head())

print("\nDataset Info")
print(df.info())

# Step 3: Define Core KPIs

# Total Revenue
total_revenue = df["Revenue"].sum()

# Average Order Value
avg_order_value = df["Revenue"].mean()

# Total Customers
total_customers = df["CustomerID"].nunique()

# Total Orders
total_orders = df["OrderID"].nunique()

print("\n----- KPI RESULTS -----")
print("Total Revenue:", total_revenue)
print("Average Order Value:", avg_order_value)
print("Total Customers:", total_customers)
print("Total Orders:", total_orders)

# Step 4: Monthly Revenue Trend
df["OrderDate"] = pd.to_datetime(df["OrderDate"])

monthly_sales = df.groupby(df["OrderDate"].dt.to_period("M"))["Revenue"].sum()

plt.figure()
monthly_sales.plot()
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()

# Step 5: Top Products by Revenue
top_products = df.groupby("ProductName")["Revenue"].sum().sort_values(ascending=False).head(10)

plt.figure()
top_products.plot(kind="bar")
plt.title("Top 10 Products by Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()

# Step 6: Customer Segmentation using KMeans

customer_data = df.groupby("CustomerID").agg({
    "Revenue":"sum",
    "OrderID":"count"
}).reset_index()

customer_data.columns = ["CustomerID","TotalSpend","TotalOrders"]

# Standardize Data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(customer_data[["TotalSpend","TotalOrders"]])

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
customer_data["Segment"] = kmeans.fit_predict(scaled_data)

print("\nCustomer Segmentation")
print(customer_data.head())

# Step 7: Visualize Segments
plt.figure()
sns.scatterplot(
    data=customer_data,
    x="TotalSpend",
    y="TotalOrders",
    hue="Segment",
    palette="deep"
)

plt.title("Customer Segmentation (K-Means)")
plt.show()

# Step 8: Save Segmented Data
customer_data.to_csv("customer_segments.csv", index=False)

print("\nAnalysis Completed. Segmented data saved.")
```

```python
# TASK 3 - Deep Dive Analysis & Dashboard Preparation

# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Step 2: Load Dataset
df = pd.read_csv("cleaned_data.csv")

print("Dataset Preview")
print(df.head())

print("\nDataset Info")
print(df.info())

# Step 3: Define Core KPIs

# Total Revenue
total_revenue = df["revenue"].sum()

# Average Order Value
avg_order_value = df["revenue"].mean()

# Total Customers
total_customers = df["Customer_ID"].nunique()

# Total Orders
total_orders = df.shape[0] # Assuming each row is a unique order since OrderID is not present

print("\n----- KPI RESULTS -----")
print("Total Revenue:", total_revenue)
print("Average Order Value:", avg_order_value)
print("Total Customers:", total_customers)
print("Total Orders:", total_orders)

# Step 4: Monthly Revenue Trend
df["Transaction_Date"] = pd.to_datetime(df["Transaction_Date"], format='mixed')

monthly_sales = df.groupby(df["Transaction_Date"].dt.to_period("M"))["revenue"].sum()

plt.figure()
monthly_sales.plot()
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()

# Step 5: Top Products by Revenue
top_products = df.groupby("Product_Category")["revenue"].sum().sort_values(ascending=False).head(10)

plt.figure()
top_products.plot(kind="bar")
plt.title("Top 10 Products by Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()

# Step 6: Customer Segmentation using KMeans

customer_data = df.groupby("Customer_ID").agg(
    TotalSpend=('revenue', 'sum'),
    TotalOrders=('Transaction_Amount', 'count') # Count Transaction_Amount to get total orders per customer
).reset_index()

# The columns are already correctly named after named aggregation and reset_index().
# No need for: customer_data.columns = ["Customer_ID","TotalSpend","TotalOrders"]

# Standardize Data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(customer_data[["TotalSpend","TotalOrders"]])

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
customer_data["Segment"] = kmeans.fit_predict(scaled_data)

print("\nCustomer Segmentation")
print(customer_data.head())

# Step 7: Visualize Segments
plt.figure()
sns.scatterplot(
    data=customer_data,
    x="TotalSpend",
    y="TotalOrders",
    hue="Segment",
    palette="deep"
)

plt.title("Customer Segmentation (K-Means)")
plt.show()

# Step 8: Save Segmented Data
customer_data.to_csv("customer_segments.csv", index=False)

print("\nAnalysis Completed. Segmented data saved.")
```

```python
# Step 1: Import Libraries
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Step 2: Load Dataset
df = pd.read_csv("cleaned_data.csv")

# Step 3: Create Revenue Column (if not already present)
if "Revenue" not in df.columns:
    df["Revenue"] = df["Price"] * df["Quantity"]

# Step 4: Create Customer Summary
customer_data = df.groupby("CustomerID").agg({
    "Revenue": "sum",
    "OrderID": "count"
}).reset_index()

customer_data.columns = ["CustomerID", "TotalSpend", "TotalOrders"]

print(customer_data.head())

# Step 5: Scale Data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(customer_data[["TotalSpend", "TotalOrders"]])

# Step 6: Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
customer_data["Segment"] = kmeans.fit_predict(scaled_data)

# Step 7: Save CSV File
customer_data.to_csv("customer_segments.csv", index=False)

print("customer_segments.csv file created successfully!")
```
